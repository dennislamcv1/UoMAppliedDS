### Blood Donations 

**Attribute Information:**

R (Recency - months since last donation)<br>
F (Frequency - total number of donation)<br>
M (Monetary - total blood donated in c.c.)<br>
T (Time - months since first donation)<br>
A binary variable representing whether he/she donated blood in March 2007 (1 stand for donating blood; 0 stands for not donating blood)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import imshow

import seaborn as sns
```

```python
df = pd.read_csv("transfusiondata.csv")
```

```python
df
```

```python
df.describe()
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,5))
plt.imshow(X=df.corr(),cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(data=df.corr(), cmap="coolwarm", annot=True,linewidths=2)
plt.title("Heatmap for correlation")
plt.show()
```

**Notes:**<br>
Comparing to Donated column, there is not strong correlation seen from heatmap.<br>
There is strong correlation between Time versus Frequency and Monetary<br>
Perfect correlation between Frequency and Monetary

```python
fig = plt.figure(figsize=(25,10))

sns.barplot(x=df.Time,y=df.Monetary, hue=df.Donated,ci=None)
plt.legend(title='Donated',loc='upper left')
plt.title("Total blood donation over time")
plt.savefig("1.png",dpi=fig.dpi)
plt.show()
```

```python
fig = plt.figure(figsize=(25,10))
sns.barplot(x=df.Time,y=df.Frequency, hue=df.Donated,ci=None)
plt.title("Total number of donation over time")
plt.legend(title='Donated',loc='upper left')
plt.savefig("2.png",dpi=fig.dpi)
plt.show()
```

```python
fig = plt.figure(figsize=(25,10))
sns.barplot(x=df.Time,y=df.Recency, hue=df.Donated,ci=None)
plt.legend(title='Donated',loc='upper left')
plt.title("Recency over time")
plt.savefig("3.png",dpi=fig.dpi)
plt.show()
```

```python
#Plot line charts
plt.figure(figsize=(25,5))
plt.plot(df.Time,df.Recency, color='blue')
plt.plot(df.Time,df.Frequency,'r')
plt.plot(df.Time,df.Monetary,'y',linestyle='dashed')
plt.show()
```
